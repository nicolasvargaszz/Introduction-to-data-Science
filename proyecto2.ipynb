## **Pandas Avanzado**

```python
import numpy as pd
import pandas as pd
import os
```

```python
os.listdir()
```

```python
archivos = []

for documento in os.listdir():
  if documento.endswith('.xlsx'):
    if documento == 'Julio_2024-1-150.xlsx':
      archivos.append(pd.read_excel(documento, skiprows=1))
    else:
      archivos.append(pd.read_excel(documento,header=None))
```
