## **Pandas Avanzado**

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
os.listdir()
```

```python
ephc = pd.DataFrame()

for documento in os.listdir():
  if documento.endswith('.csv'):
    print(documento)
    ephc = pd.concat([ephc,pd.read_csv(documento,sep=";",low_memory=False)])

ephc
```
