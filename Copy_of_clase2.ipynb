## **Introducción a Matplotlib**

### **¿Qué es Matplotlib?**

>Matplotlib es una biblioteca de Python que permite crear gráficos y visualizaciones de datos. Fue diseñada para parecerse a MATLAB, que es una plataforma comercial utilizada ampliamente en ingeniería y ciencias para análisis de datos, visualización y cálculo numérico. A diferencia de MATLAB, Matplotlib es completamente gratuita y de código abierto.

### **Importancia y aplicaciones en la visualización de datos**

**Importancia de Matplotlib en el Análisis de datos**

* ***Visualización Clara:*** Matplotlib nos ayuda a visualizar nuestros datos de una manera que sea fácil de entender. Los gráficos pueden revelar patrones, tendencias y relaciones que no son evidentes en tablas de datos crudos.

* ***Flexibilidad:*** Permite crear una amplia variedad de gráficos, desde simples gráficos de líneas hasta complejas visualizaciones en 3D.

* ***Interoperabilidad:*** Se integra bien con otras bibliotecas de análisis de datos como NumPy y Pandas, lo que facilita el flujo de trabajo en análisis de datos.

* ***Personalización:*** Ofrece un alto grado de personalización, permitiéndonos ajustar cada aspecto del gráfico para que se adapte a nuestras necesidades específicas.

**Aplicaciones de Matplotlib en el Análisis de Datos**

* ***Comparación de Datos:*** Facilita la comparación entre diferentes conjuntos de datos, ayudando a entender las diferencias y similitudes.

* ***Visualización de Modelos:*** Ayuda a visualizar el rendimiento de modelos de machine learning, por ejemplo, mostrando curvas de error, matrices de confusión, etc.

### **Gráficos Anidados**

>Crear gráficos anidados, como hacer zoom en un gráfico.

````python
np.random.seed(0)

x = np.linspace(0,10,100)
y = np.sin(x) + np.random.normal(0,0.1,100)

figuras,ejes = plt.subplots(figsize=(10,6),dpi=500)
ejes.plot(x,y)

zoom = figuras.add_axes([0.5,0.47,0.35,0.35])
zoom.plot(x,y)
zoom.set_xlim(2,4)
zoom.set_ylim(0,1)

figuras.savefig("Grafico_con_zoom.png")
plt.show()
````

### **Histogramas**

#### **Histograma Básico**

>Utilidad: Muestra la distribución general de los datos.

>Ejemplo: Distribución de las edades de un grupo de personas.

````python
edades = np.random.randint(10,100,100)
plt.hist(edades,bins=20,color="green",edgecolor="black")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Histograma de Edades")
plt.show()
````

#### **Histograma Comparativo**

>Utilidad: Compara las distribuciones de dos o más conjuntos de datos.

>Ejemplo: Comparación de las alturas de hombres y mujeres.

````python
np.random.seed(0)
hombres = np.random.randint(160,200,100)
mujeres = np.random.randint(150,190,100)
plt.figure(figsize=(10,6),dpi=300)
plt.hist([hombres,mujeres],bins=20,label = ["Altura de Hombres","Altura de Mujeres"], color=["blue","pink"],edgecolor="black")
plt.xlabel("Altura")
plt.ylabel("Frcuencia")
plt.title("Histograma comparativo de alturas.")
plt.legend()
plt.show()
````

#### **Histograma Acumulativo**

>Utilidad: Muestra la acumulación de datos a lo largo de la distribución.

>Ejemplo: Porcentaje acumulativo de ventas durante el año.

````python
clase1 = np.random.randint(0,100,50)
clase2 = np.random.randint(0,80,50)

plt.hist([clase1,clase2],bins=5,hatch="/",edgecolor="Black",alpha=0.9)
plt.show()
````

#### **Histograma de Paso**

>Utilidad: Evita la superposición de múltiples histogramas, manteniendo la claridad.

>Ejemplo: Comparación de distribuciones de ingresos en diferentes regiones.

````python
import numpy as np

# Data setup
ciudades = ["Asuncion", "Luque", "San Lorenzo", "Fernando de la Mora","Ciudad del Este"]
dias = np.arange(1, 366)
dias = np.tile(dias, 5)
temperaturas = np.random.randint(10, 40, 365)
ciudades = np.repeat(ciudades, 365)
temperaturas = np.tile(temperaturas, 5)

# Verify shapes are equal
print(ciudades.shape == temperaturas.shape)  # Should be True

# Stack the arrays into one
resultado = np.column_stack((dias, ciudades, temperaturas))
print(resultado.shape)

# Save the result to a CSV file
np.savetxt("temperaturas.csv", resultado, delimiter=",", header="Dia,Ciudad,Temperatura", comments="", fmt="%s",encoding="UTF-8")
````

````python
# Cargamos nuestros datos.

datos = np.loadtxt("temperaturas.csv",delimiter=',',skiprows=1,dtype=str)
datos
````

````python
dias_cargados = datos[:,0].astype(int)
ciudades_cargadas = datos[:,1]
temperaturas_cargadas = datos[:,2].astype(float)
````

````python
ciudades_unicas = np.unique(ciudades_cargadas)
temperaturas_medias_mensuales = {ciudad:[] for ciudad in ciudades_unicas}
for ciudad in ciudades_unicas:
  for mes in range(1,13):
    mask = (ciudades_cargadas == ciudad) & ((dias_cargados - 1)//30 == mes - 1)
    temperaturas_media = np.mean(temperaturas_cargadas[mask])
    temperaturas_medias_mensuales[ciudad].append((mes,temperaturas_media))
````

````python
temperaturas_medias_mensuales
````

````python
import matplotlib.pyplot as plt

# Assuming ciudades_unicas and temperaturas_medias_mensuales are defined
plt.figure(figsize=(12, 6), dpi=300)

for ciudad in ciudades_unicas:
    meses, temperaturas = zip(*temperaturas_medias_mensuales[ciudad])
    plt.plot(meses, temperaturas, label=ciudad)

# Correct function usage
plt.xlabel = "mes"
plt.ylabel("Temperatura media")
plt.title("Temperatura media mensual por Ciudad")
plt.legend()  # Add legend to the plot

# Set x-axis labels for months
plt.xticks(range(1, 13), ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"])

plt.show()
````
